## terms


Here "*this context*" refers to the 2019 drifter/glider experiment in the southern ocean.


* FSLE is Finite Size Lyapunov Exponent $\lambda$, a (spectral) parameter characterizing the rate of separation of 'infinitely close' trajectories. 
[The data product originates at Aviso.](https://www.aviso.altimetry.fr/en/data/products/value-added-products/fsle-finite-size-lyapunov-exponents/fsle-description.html)
Because it is spectral: The LE includes a maximum which, if greater than zero, indicates the system is chaotic. In this context it is a time-evolving map of turbulent structure scale.


* A front is the interface of two fluids. 
    * Polar Front abbreviated **PF** is a large-scale (200km width; 2 deg latitude; mesoscale)
transitional region in the southern ocean. Figure 4 (a) and (b) at 30 deg E show lateral transition in T/S profile
across 50 -- 52 deg S.
    * **PF core** refers to water within the PF.


* ADT is **Absolute Dynamic Topography** and is somewhat unresolved for me at the moment. It has paramount emphasis 
in the notes. 
[Here is a resource](https://cmr.earthdata.nasa.gov/search/concepts/C1214586177-SCIOPS.html) 
associated with CMR. 


* The wind rose in Figure 2 (c) follows the oceanographic convention of "from-to", opposite meteorology. 
The winds are blowing from the west / west northwest as expected.


* The $N^2$ parameter from Figure 4 (c) is technically *buoyancy frequency* or *Brunt-Vaisala frequency*. The larger it is the more stable the water mass; in this case the water below the mixed layer. 


* AOU is oxygen deficit: Dissolved oxygen observed in relation to capacity (varies with physical parameters).
Is there a conceptually related **UPD**?


* Conservative Temperature $T_{C}$ also abbreviated $\Theta$: Zeroth order a slightly modified version
of observed temperature. First order hand-waving: My background suggests salinity is 
easily conservative in mixing (example: estuaries) and the idea applied to temperature gets a bit more
involved because of salinity and temperature dependence of heat capacity of water. 
Best definition I saw: It is the temperature for which, upon mixing two equal water masses 
with respective $\Theta_{1},\;\Theta_{2}$ we get a new water mass with conservative temperature 
equal to their mean.

subduction as displacement.
