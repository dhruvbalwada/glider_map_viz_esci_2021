```
# define GP model
class ExactGPModel(gpytorch.models.ExactGP):
    
    def __init__(self, train_x, train_y, likelihood):
        super(ExactGPModel, self).__init__(train_x, train_y, likelihood)
        self.mean_module = gpytorch.means.ConstantMean()
        self.covar_module = gpytorch.kernels.ScaleKernel(gpytorch.kernels.MaternKernel(nu=1.5, ard_num_dims=train_x.shape[1]))

    def forward(self, x):
        mean_x = self.mean_module(x)
        covar_x = self.covar_module(x)
        return gpytorch.distributions.MultivariateNormal(mean_x, covar_x)
```

```
# data container
class SOSEDataset(Dataset):

  def __init__(self, train_x, train_y, time):
    self._train_x = train_x
    self._train_y = train_y
    self._time = time

  def __len__(self):
    return 5

  def __getitem__(self, idx):
    start, end = None, None
    if idx == 0:
      start = (datetime.strptime('20140101', '%Y%m%d') - datetime(1970, 1, 1)).days
      end = (datetime.strptime('20141231', '%Y%m%d') - datetime(1970, 1, 1)).days
    elif idx == 1:
      start = (datetime.strptime('20150101', '%Y%m%d') - datetime(1970, 1, 1)).days
      end = (datetime.strptime('20151231', '%Y%m%d') - datetime(1970, 1, 1)).days
    elif idx == 2:
      start = (datetime.strptime('20160101', '%Y%m%d') - datetime(1970, 1, 1)).days
      end = (datetime.strptime('20161231', '%Y%m%d') - datetime(1970, 1, 1)).days
    elif idx == 3:
      start = (datetime.strptime('20170101', '%Y%m%d') - datetime(1970, 1, 1)).days
      end = (datetime.strptime('20171231', '%Y%m%d') - datetime(1970, 1, 1)).days
    elif idx == 4:
      start = (datetime.strptime('20180101', '%Y%m%d') - datetime(1970, 1, 1)).days
      end = (datetime.strptime('20181231', '%Y%m%d') - datetime(1970, 1, 1)).days
    else:
      raise IndexError('dataset index out of bound.')
    return self._train_x[(start < self._time) & (self._time < end), :], self._train_y[:, (start < self._time) & (self._time < end)]
```

```
def get_window(needle, lat, lon, window):
    """Find indices of lat-lon points that are within the window centered at needle
    """
    EARTH_RADIUS_METERS = 6366565.0
    dlat = np.radians(lat) - np.radians(needle[0])
    dlon = np.radians(lon) - np.radians(needle[1])
    a = np.square(np.sin(dlat / 2.0)) + np.cos(np.radians(needle[0])) * np.cos(np.radians(lat)) * np.square(np.sin(dlon / 2.0))
    great_circle_distance = 2 * np.arcsin(np.minimum(np.sqrt(a), np.repeat(1, len(a))))
    d = EARTH_RADIUS_METERS * great_circle_distance
    return d < window
```

```
# fit GP models
def univ_gp_all_data(window_size,
                     data,
                     var_name,
                     lat_grid,
                     lon_grid,
                     depth,
                     verbose=True):

    date = data.prof_YYYYMMDD
    lon = data.prof_lon
    lat = data.prof_lat
    var = data[var_name]

    lon_grid_mesh, lat_grid_mesh = np.meshgrid(lon_grid, lat_grid)
    ocean_mask = maskoceans(lon_grid_mesh,
                            lat_grid_mesh,
                            np.zeros([len(lat_grid),
                                      len(lon_grid)]),
                            inlands=False,
                            grid=1.25).mask


    for ipres, pres_lvl in enumerate(depth):
        var_ = var.sel(iDEPTH=pres_lvl)
        
        mask = var_.notnull()
        
        var_ = var_.where(mask, drop=True)

        

        t_ = date.where(mask, drop=True)
        lat_ = lat.where(mask, drop=True)
        lon_ = lon.where(mask, drop=True)
        lon_[lon_ > 180] = lon_[lon_ > 180] - 360


        t_ = np.array([(datetime.strptime(str(int(x)), '%Y%m%d') - datetime(1970, 1, 1)).days
              for x in t_])
        lat_ = lat_.values
        lon_ = lon_.values
        var_ = var_.values

        # function for parallel computing
        ########################################################################
        @ray.remote
        def parallel_gp(lat_, lon_, t_, var_, lat_grid_, lon_grid_, ocean_mask, lat_indices):
          lat_grid = lat_grid_[lat_indices[0]: lat_indices[1]]
          lon_grid = lon_grid_

          # dict for model
          models = {}

          for ilat in range(len(lat_grid)):
            
            for ilon in range(len(lon_grid)):
              
              if not ocean_mask[ilat, ilon]:
                continue

              idata = get_window((lat_grid[ilat], lon_grid[ilon]), lat_,
                                lon_, window_size)   # .values

              # if no enough data around, output None for estimate
              if sum(idata) < 30:
                  continue

              print(sum(idata))
              var_tmp = var_[idata]
              lat_tmp, lon_tmp = lat_[idata], lon_[idata]
              t_tmp = t_[idata]

              lat_tmp = lat_tmp - lat_grid[ilat]
              lon_tmp = lon_tmp - lon_grid[ilon]
              
              lat_tmp = torch.tensor(lat_tmp, dtype=torch.float)
              lon_tmp = torch.tensor(lon_tmp, dtype=torch.float)
              t_tmp = torch.tensor(t_tmp, dtype=torch.float)
              var_tmp = torch.reshape(torch.tensor(var_tmp, dtype=torch.float), (-1, 1))
              train_x = torch.transpose(torch.vstack((lat_tmp, lon_tmp, t_tmp)), 0, 1)
              train_y = torch.transpose(var_tmp, 0, 1)
              
              
              likelihood = gpytorch.likelihoods.GaussianLikelihood()
              model = ExactGPModel(train_x, train_y, likelihood)

              
              model.train()
              likelihood.train()


              # optimizer = torch.optim.LBFGS(model.parameters(), 
              #                               lr=0.1, 
              #                               max_iter=10) 


              optimizer = torch.optim.Adam(model.parameters(), lr=0.1)

              # the marginal log likelihood
              mll = gpytorch.mlls.ExactMarginalLogLikelihood(likelihood, model)

              # need this for LBFGS
              # def closure():
              #   output = model(train_x)
              #   optimizer.zero_grad()
              #   loss = -mll(output, train_y)
              #   loss.backward()
              #   # print('Iter %d/%d - Loss: %.3f   lengthscale: %s   noise: %.3f' % (
              #   #     i + 1, training_iter, loss.item(),
              #   #     str(model.covar_module.base_kernel.lengthscale),
              #   #     model.likelihood.noise.item()
              #   # ))
              #   return loss 
              # optimizer.step(closure)

              num_iter = 20
              for i in range(num_iter):
                start = time.time()
                optimizer.zero_grad()
                output = model(train_x)
                loss = -mll(output, train_y)
                l = loss.item()
                mid = time.time()
                loss.backward()
                step = time.time()               
                optimizer.step()
                print(f'loss: {l}\tTime for this iteration: {time.time() - start}\tloss calculation: {mid - start}\tstep: {time.time() - step}\tloss backward: {step - mid}')

              key = str(lat_grid[ilat]) + '_' + str(lon_grid[ilon])
              models[key] = model.state_dict()
          
          return models
        ########################################################################

        #  for parallel computing
        ray.init(num_cpus=2, ignore_reinit_error=True)
        lat_id = ray.put(lat_)
        lon_id = ray.put(lon_)
        t_id = ray.put(t_)
        var_id = ray.put(var_)
        lat_grid_id = ray.put(lat_grid)
        lon_grid_id = ray.put(lon_grid) 
        ocean_mask_id = ray.put(ocean_mask)


        lat_indices_list = [(0, int(len(lat_grid) // 2)), (int(len(lat_grid) // 2), len(lat_grid))]
        futures = []
        for tup in lat_indices_list:
          futures.append(parallel_gp.remote(lat_id, lon_id, t_id, var_id, lat_grid_id, lon_grid_id, ocean_mask_id, tup))
        res_ = ray.get(futures)
        

        models_dict = {k: v for d in res_ for k, v in d.items()}
        ray.shutdown()


        if verbose:
            print(f'{ipres}th depth done.')
            
    return models_dict
           
```

```
def inference(models_dict, data, var_name, lat_grid, lon_grid, depth, day='20160101', day_window=100, geo_window=0.5*1e6):
  
  date = data.prof_YYYYMMDD
  lon = data.prof_lon
  lat = data.prof_lat
  var = data[var_name]

  estim_var = np.full((len(depth), len(lat_grid), len(lon_grid)), fill_value=np.nan)
  var_var = np.full((len(depth), len(lat_grid), len(lon_grid)), fill_value=np.nan)
  noise_var = np.full((len(depth), len(lat_grid), len(lon_grid)), fill_value=np.nan)
  scale_var = np.full((len(depth), len(lat_grid), len(lon_grid)), fill_value=np.nan)
  lengthscale_var = np.full((len(depth), len(lat_grid), len(lon_grid), 3), fill_value=np.nan)

  lon_grid_mesh, lat_grid_mesh = np.meshgrid(lon_grid, lat_grid)
  ocean_mask = maskoceans(lon_grid_mesh,
                          lat_grid_mesh,
                          np.zeros([len(lat_grid),
                                    len(lon_grid)]),
                          inlands=False,
                          grid=1.25).mask
  
  for ipres, pres_lvl in enumerate(depth):
    
    var_ = var.sel(iDEPTH=pres_lvl)
    
    date_mask = xr.apply_ufunc(lambda x: abs(pd.to_datetime(x, format='%Y%m%d') - datetime.strptime(day, '%Y%m%d')).days < day_window, 
                               date)
    mask = var_.notnull() * date_mask
    var_ = var_.where(mask, drop=True)
    
    t_ = date.where(mask, drop=True)
    lat_ = lat.where(mask, drop=True)
    lon_ = lon.where(mask, drop=True)
    lon_[lon_ > 180] = lon_[lon_ > 180] - 360

    t_ = np.array([(datetime.strptime(str(int(x)), '%Y%m%d') - datetime(1970, 1, 1)).days
          for x in t_])
    lat_ = lat_.values
    lon_ = lon_.values
    var_ = var_.values

    for ilat in range(len(lat_grid)):
      
      for ilon in range(len(lon_grid)):
        
        if not ocean_mask[ilat, ilon]:
          continue
        
        idata = get_window(((lat_grid[ilat], lon_grid[ilon])), lat_,
                          lon_, geo_window)

        if sum(idata) < 30:
                  continue

        var_tmp = var_[idata]
        lat_tmp, lon_tmp = lat_[idata], lon_[idata]
        t_tmp = t_[idata]

        lat_tmp = lat_tmp - lat_grid[ilat]
        lon_tmp = lon_tmp - lon_grid[ilon]
        
        lat_tmp = torch.tensor(lat_tmp, dtype=torch.float)
        lon_tmp = torch.tensor(lon_tmp, dtype=torch.float)
        t_tmp = torch.tensor(t_tmp, dtype=torch.float)
        var_tmp = torch.reshape(torch.tensor(var_tmp, dtype=torch.float), (-1, 1))
        train_x = torch.transpose(torch.vstack((lat_tmp, lon_tmp, t_tmp)), 0, 1)
        train_y = torch.transpose(var_tmp, 0, 1)
        
        
        likelihood = gpytorch.likelihoods.GaussianLikelihood()
        model = ExactGPModel(train_x, train_y, likelihood)

        # load model params
        model.load_state_dict(models_dict[str(lat_grid[ilat]) + '_' + str(lon_grid[ilon])])

        model.eval()
        likelihood.eval()

        with torch.no_grad(), gpytorch.settings.fast_pred_var():
          # use number of days since 1970
          days = (datetime.strptime(day, '%Y%m%d') - datetime(1970, 1, 1)).days
          est_x = torch.tensor([[0, 0, days]], dtype=torch.float)
          
          observed_pred = likelihood(model(est_x))
          estim_var[ipres, ilat, ilon] = observed_pred.mean.item()
          var_var[ipres, ilat, ilon] = observed_pred.variance.item()
          scale_var[ipres, ilat, ilon] = model.covar_module.outputscale.item()
          noise_var[ipres, ilat, ilon] = model.likelihood.noise.item()
          lengthscale_var[ipres, ilat, ilon, :] = model.covar_module.base_kernel.lengthscale.detach().numpy()

  return {'mean': estim_var, 'variance': var_var, "noise": noise_var, "scale": scale_var, "lengthscale": lengthscale_var}
```

```
from matplotlib.patches import Rectangle

def plot_anom(anom, var, raw=False, second_order=False):
  plt.rc('font', size=20)
  fig, ax = plt.subplots(figsize=(20, 15))
  ax = plt.axes(projection=ccrs.PlateCarree())
  ax.set_extent([-180, -60, -70, -40], crs=ccrs.PlateCarree())
  ax.stock_img()
  gl = ax.gridlines(crs=ccrs.PlateCarree(), draw_labels=True, 
    linewidth=1.2, color='k', alpha=0., linestyle='--')
  gl.top_xlabels = False  
  gl.top_ylabels = False  
  gl.xformatter = LONGITUDE_FORMATTER 
  gl.yformatter = LATITUDE_FORMATTER 

  ct = ax.scatter(x=anom.prof_lon.where(anom[var][:, 0].notnull(), drop=True), y=anom.prof_lat.where(anom[var][:, 0].notnull(), drop=True),
          c=anom[var][:, 0].where(anom[var][:, 0].notnull(), drop=True), marker='o', alpha=1, s=20,
          transform=ccrs.PlateCarree(),
          )
  ax.add_patch(Rectangle((-130 - .1, -50 - .1), 20, 5, fill=None, alpha=1))
  raw = 'raw' if raw else ''
  second_order = 'second order' if second_order else ''
  fig.colorbar(ct, fraction=0.01)
  var = 'temperature' if var == 'anom_T' else 'salinity'
  ax.set_title(f'${var}$ anomalies at depth = 200m', fontsize=20)
  plt.show()


def plot_gp(data, var, param):
  plt.rc('font', size=20)
  fig, ax = plt.subplots(figsize=(20, 15))
  ax = plt.axes(projection=ccrs.PlateCarree())
  ax.set_extent([-180, -60, -70, -40], crs=ccrs.PlateCarree())
  ax.stock_img()
  gl = ax.gridlines(crs=ccrs.PlateCarree(), draw_labels=True, 
    linewidth=1.2, color='k', alpha=0., linestyle='--')
  gl.top_xlabels = False  
  gl.top_ylabels = False  
  gl.xformatter = LONGITUDE_FORMATTER 
  gl.yformatter = LATITUDE_FORMATTER 

  x = np.arange(-130., -110., 1)
  y = np.arange(-50., -45., 1)


  x, y = np.meshgrid(x, y)

  ct = ax.scatter(x=x, y=y,
          c=data[param][0], marker='o', alpha=1, s=100,
          transform=ccrs.PlateCarree(),
          )

  fig.colorbar(ct, fraction=0.01)
  ax.set_title(f'${var}$ anomalies {param} at depth = 200m', fontsize=20)
  plt.show()
  
```
